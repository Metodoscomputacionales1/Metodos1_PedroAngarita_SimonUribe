```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 #MARKOV

T=np.array([[0.3,0.25,0.15,0.1],\
            [0.3,0.25,0.55,0.2],\
            [0.3, 0.5, 0.1,0.3],\
            [0.1,0.,0.2,0.4]])
```

```python
np.sum(T,axis=0)
```

```python
A=np.linalg.matrix_power(T,2) #cantidad de potencias
A
```

```python
initial=np.array([1,0,0,0])
Res=np.dot(A,initial)
#Probabilidad de trancicion 1 a 4
Res[3]
```

```python
C1=[]
C2=[]
C3=[]
C4=[]
for i in range(10):
  A=np.linalg.matrix_power(T,i)
  resultado=np.dot(A,initial)
  C1.append(resultado[0])
  C2.append(resultado[1])
  C3.append(resultado[2])
  C4.append(resultado[3])

plt.plot(C1,label='C1')
plt.plot(C2,label='C2')
plt.plot(C3,label='C3')
plt.plot(C4,label='C4')
plt.xlabel('Transicion')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```
