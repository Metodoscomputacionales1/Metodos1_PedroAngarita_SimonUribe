```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=365
sum=1
y=[]
x=np.linspace(1,80,80)
for i in range(80):
    sum*=(n-i)/365
    y.append(sum)
plt.scatter(x,y)
```
