```python
# Trayectoria de bala V0 y 0

import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

x=sym.Symbol("x", real=True)

def lagrange(x,X,i):

  L = 1
  for j in range(len(X)):
    if i != j:
      L *= (x-X[j])/(X[i]-X[j])
  return L


def interpolar(x,X,Y):

  Poly = 0
  for i in range(len(X)):
    Poly += Y[i]*lagrange(x,X,i)
  return Poly

X=np.array([1.4, 3.5, 5.6])
Y=np.array([0.40079549, 0.5941281, 0.29802796])
plt.scatter(X,Y,color="r")

L0=lagrange(x,X,0)
L1=lagrange(x,X,1)
L2=lagrange(x,X,2)


yx=interpolar(x,X,Y).expand()
x0=np.linspace(X[0],X[-1],100)
y0=interpolar(x0,X,Y)

plt.plot(x0,y0)
coeficientes=np.polyfit(x0,y0, 2)
tCuadratico=coeficientes[0]
tLineal=coeficientes[1]

print(yx)
angulo=np.arctan(tLineal) # en radianes
V0=((-1)*(9.8/(tCuadratico*2*(np.cos(angulo)**2))))**(1/2)
print("Velocidad inicial: " + str(V0))
angulo_grados=angulo*180/np.pi
print("Angulo en grados: " + str(angulo_grados) + "Â°")
```
