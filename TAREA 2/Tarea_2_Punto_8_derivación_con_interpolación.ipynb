```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
#conjunto soporte raiz de tangnete/h=0.01
X=np.linspace(0.1,1.1,101)
Y=np.sqrt(np.tan(X) )
plt.scatter(X,Y)
```

```python
#calculo polinomio interpola conjunto soporte
x = sym.Symbol('x',real=True)
def Lagrange(x,X,i):

  L = 1
  for j in range(len(X)):
    if i != j:
      L *= (x-X[j])/(X[i]-X[j])
  return L
```

```python
#prueba primer nodo del conjunto
Lagrange(x,X,0)
```

```python
def Interpolate(x,X,Y):

  Poly = 0
  for i in range(len(X)):
    Poly += Y[i]*Lagrange(x,X,i)
  return Poly
```

```python
x1 = np.linspace(X[0],X[-1],101)
y1 = Interpolate(x1,X,Y)
```

```python
# grafica funcion interpolada

plt.plot(x1,y1)
```

```python
#simplificacion del la funcion interpolada
poly = Interpolate(x,X,Y)
poly=sym.expand(poly)
poly
```

```python
#derivada polimio interpolador
def derivada(poly):
  return sym.diff(poly,x)
derivada(poly)
```

```python
#derivada numerica progresiva
def derivada_progresiva(x,y):
  h=x[1]-x[0]


  return(1/(2*h))*((-3*y[:-2])+(4*y[1:-1])-y[2:])
```

```python
a=derivada_progresiva(x1,y1)
plt.plot(x1[1:-1],a)
```

```python
# derivada central de la interpolacion
def derivada_central(x,y):
  h=x[1]-x[0]
  print(h)
  return(1/(2*h))*(y[2:]-y[:-2])
plt.plot(x1[1:-1],derivada_central(x1,y1))
```

```python
#derivada analitica
def derivada_analitica(x):
  return np.square(1/np.cos(x))*(1/(2*(np.sqrt(np.tan(x)))))
plt.plot(x1[1:-1],derivada_analitica(x1[1:-1]))
```

```python
#error nodal derivada progresiva de orden dos
plt.plot(x1[1:-1],np.abs(a-derivada_analitica(x1[1:-1])))
#
```

```python
# error nodal derivada cental
plt.plot(x1[1:-1],np.abs(derivada_central(x1,y1)-derivada_analitica(x1[1:-1])))
```

```python
## el error aumenta en gran cantidad cuando el punto nodal esta cerca del 0,
# asi mismo podemos observar como la derivada central es mucho mas exacta, aun asi mantengan el mismo orden
#no no tienen en mismo orden la derivada central tiene un orden de h al cuadrado
# mientras que por algun motivo la formula propuesta mantiene un orden de h
```
