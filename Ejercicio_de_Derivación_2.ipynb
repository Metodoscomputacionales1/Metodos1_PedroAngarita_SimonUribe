```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
nodos=np.linspace(-10,10,401
                  )

```

```python
def funcion(x:float):
  return 1/((1+1/(np.exp((x**2))))**(1/2))
funcion(-10)
```

```python
def funcion_derivada_real(x:float):
  return np.exp(-x**2)*x/((1+np.exp(-x**2))**(3/2))
derivadas_reales=[]
for i in nodos:
  derivadas_reales.append(funcion_derivada_real(i))
plt.plot(nodos,derivadas_reales)
```

```python
imagenes=[]
for a in nodos:
  imagenes.append(funcion(a))

plt.plot(nodos,imagenes)
```

```python

def derivada_central_por_nodo(x:list):
  h=0.05
  derivada=[]
  for i in range(1,len(x)-1):
    derivada.append((x[i+1]-x[i-1])/(2*h))

  return derivada
derivadas_centrales=derivada_central_por_nodo(imagenes)
nodos2=nodos[1:-1]
plt.plot(nodos2,derivadas_centrales)
```

```python
#error_local
errror_local=[]
derivadas_reales2=derivadas_reales[1:-1]
for i in range(len(derivadas_reales2)):
  errror_local.append(abs(derivadas_reales2[i]-derivadas_centrales[i]))
```
