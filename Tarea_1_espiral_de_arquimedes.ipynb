```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
linespace=np.linspace(0,6.28318530718,100)
```

```python
x=[]
y=[]
for i in linespace:
    x.append(np.cos(i)*i)
    y.append(np.sin(i)*i)

plt.plot(x,y)


```
