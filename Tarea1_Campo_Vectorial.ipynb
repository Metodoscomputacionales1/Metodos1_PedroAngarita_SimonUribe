```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-4,4,25)
y=np.linspace(-4,4,25)
X,Y=np.meshgrid(x,y)

h=0.001

def potencial(X,Y,R,V):
    A= V*X*(1-((R**2)/((X**2)+(Y**2))))
    return A

def derivX (X,Y, h):
    D=(potencial((X+h),Y,2,2)-potencial((X-h),Y,2,2))/(2*h)
    L=X**2+Y**2
    D[L<=4]=0
    return D

def derivY (X,Y, h):
    D=(-1)*(potencial(X,(Y+h),2,2)-potencial(X,(Y-h),2,2))/(2*h)
    L=X**2+Y**2
    D[L<=4]=0
    return D


fig = plt.figure()
ax = plt.subplot()
ax.quiver(X,Y,derivX(X,Y,h),derivY(X,Y,h),color="blue")
c=plt.Circle((0,0),2, fill=False, color="orange")
ax.add_patch(c)
```
