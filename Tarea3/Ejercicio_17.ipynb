```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
#a)
Raices, Pesos=np.polynomial.laguerre.laggauss(3)
Raices, Pesos

x=sym.Symbol("x", real=True)

f= lambda x:  x**3 / (1 - np.exp(-x) )

real=(np.pi**4)/15

I=np.sum(Pesos*f(Raices))
print(I)
print(real)
```

```python
#b)
X=np.linspace(2,10,9)
errores=[]
for i in range(2, 11):
    Raices, Pesos=np.polynomial.laguerre.laggauss(i)
    I=np.sum(Pesos*f(Raices))
    error=I/real
    errores.append(error)
    plt.scatter(i,error)
plt.plot(X,errores)
```
