```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
#10A
```

```python
h = sym.Symbol('h',real=True)
h
a = sym.Symbol('a',real=True)
b = sym.Symbol('b',real=True)
```

```python
X=np.array ([0,h,2*h,3*h])
```

```python

x = sym.Symbol('x',real=True)
x
```

```python

def cardinales(x,X,i):

  L = 1
  for j in range(len(X)):
    if i != j:
      L *= (x-X[j])/(X[i]-X[j])
  return L
```

```python
c1=cardinales(x,X,0)
c1=sym.expand(c1)
c1
```

```python
c2=cardinales(x,X,1)
c2=sym.expand(c2)
c2
```

```python
c3=cardinales(x,X,2)
c3=sym.expand(c3)
c3
```

```python
c4=cardinales(x,X,3)
c4=sym.expand(c4)
c4
```

```python
I1=sym.integrate(c1,(x,0,3*h))

T1=I1.subs(b - a, 3*h)
T1=I1.subs(a- b, -3*h)
T1=sym.simplify(T1)
T1
```

```python
I2=sym.integrate(c2,(x,0,3*h))
I2=sym.simplify(I2)
T2=I2.subs((b - a), 3*h)
T2=sym.simplify(T2)

T2
```

```python
I3=sym.integrate(c3,(x,0,3*h))
I3=sym.simplify(I3)
T3=I3.subs(b - a, 3*h)
T3=sym.simplify(T3)
T3
```

```python
I4=sym.integrate(c4,(x,0,3*h))
I4=sym.simplify(I4)
T4=I4.subs(b - a, 3*h)
T4=sym.simplify(T4)
T4
```

```python
#como vemos al integral de 0 a 3h las funciones cardinales nos da exactamente
#los coeficientes de f(a)+f((2a+b)/3)+f((a+2b)/3) y f(b)// o f(0)+f(h)+f(2h)+f(3h)
#por lo tanto la formula queda demostrada
```
