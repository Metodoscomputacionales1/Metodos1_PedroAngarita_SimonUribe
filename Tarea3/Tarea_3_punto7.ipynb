```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
R=1
n=2000

x=np.linspace(-R,R,n+1)
y=np.linspace(-R,R,n+1)
X,Y=np.meshgrid(x,y)
X
#area_cuadraditos
ac=np.square(2/n)
X
```

```python
y
```

```python
plt.scatter(X,Y)
```

```python

def media_esfera(x,y):
    return np.sqrt(-(x**2)-y**2+R**2)
```

```python
F0 = np.ones_like(X)

for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    F0[i,j] = media_esfera(X[i,j],Y[i,j])




F0=np.nan_to_num(F0)

F0
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,F0)
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Surface(z=F0)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.show()
```

```python
volumen = 0.0

#volumen media esfera
for i in range(len(x) - 1):
    for j in range(len(y) - 1):
        alt= (F0[j, i] + F0[j+1, i] + F0[j, i+1] + F0[j+1, i+1]) / 4.0

        volumen += alt* ac
volumen
```

```python
#valor real media esfera radio uno
a=np.pi * (((R**3)/3))*2
a
```

```python
#error con 2000 nodos
e=abs(volumen-a)
e
```
