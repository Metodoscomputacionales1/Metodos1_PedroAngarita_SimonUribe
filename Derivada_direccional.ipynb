```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f2(x,y):
  return x**4 + y**4 - 2*(x-y)**2
```

```python
Dx= lambda x,y, h=1e-5: (f2(x+h, y)-f2(x-h,y))/(2*h)
Dy= lambda x,y, h=1e-5: (f2(x, y+h)-f2(x,y-h))/(2*h)
```

```python
x0,y0=0.5,0.2
```

```python
D=lambda x, y : np.array([Dx(x,y), Dy(x,y)])
```

```python
D(x0,y0)
```

```python
theta=np.arctan(D(x0,y0)[1]/D(x0,y0)[0])+np.pi
```

```python
theta=theta*180/np.pi
theta
```

```python
u= lambda theta: np.array([np.cos(theta), np.sin(theta)])
theta=np.pi
u(theta)
```

```python
plt.arrow(x0,y0,D(x0,y0)[0],D(x0,y0)[1], color="red", head_width=0.05)
plt.arrow(x0,y0,u(theta)[0],u(theta)[1], color="blue", head_width=0.05)
```

```python
#Calculo de derivada direccional
def derivdirect (x0,y0, theta):
  return np.dot(D(x0,y0), u(theta))

derivdirect(x0,y0,theta)
#Cambio de la funcion en direcion theta
```

```python
derivdirect(x0,y0,np.pi/2)
```

```python
thetagrad=np.arctan(D(x0,y0)[1]/D(x0,y0)[0])+np.pi
derivdirect(x0,y0,thetagrad)
```
